```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
```

```python
gender = Series([50, 39], index=['male', 'female'])
```

```python
plt.rcParams['font.sans-serif'] = ['simhei']
plt.rcParams['axes.unicode_minus'] = False
#
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
ax[0].bar(gender.index, gender)
ax[0].set_title('男女比例(未优化)')
ax[1].bar(gender.index, gender)
ax[1].set_title('男女比例(优化后)')
sns.despine(ax=ax[1], left=True, bottom=True)
ax[1].set_yticks([])
for x, y in zip(gender.index, gender):
    txt = f'{y}%'
    plt.text(x, y-5, txt, fontsize=30, ha="center", va="bottom", color="w")
plt.xticks(fontsize=20)
plt.yticks([])
plt.subplots_adjust(wspace=.05);
```
